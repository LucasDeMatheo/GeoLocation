```python
import pandas as pd
#pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from geopy.geocoders import Nominatim
```

```python
# Importando os dados

file = r'tabela6910_Completa.xlsx'
cidades = pd.read_excel(file)
cidades.head(10)
```

```python
cidades.shape
```

```python
# Gerando a lista com "Ebdereço" completo

regioesEncontradas = []
for indice, linha in cidades.iterrows():
    #nome_local = linha.Regiao 
    nome_local = linha.Regiao + ', ' + linha.Estado + ', ' +  linha.Pais
    nome_local = nome_local
    regioesEncontradas.append(nome_local)
```

```python
# Criando uma coluna com as regiões encontradas
# Possibilitando a busca em dicionário pelo "endereço" completo - Resolvendo bug de mais de uma localização com mesmo nome 

cidades['RegiaoComp'] = regioesEncontradas
cidades.tail(10)
```

```python
regioesEncontradas = np.unique(regioesEncontradas)
regioesEncontradas
```

```python
regioesEncontradas.shape
```

```python
# Utilizando a API

# instantiate a new Nominatim client
app = Nominatim(user_agent="tutorial")

i=0
latitude = []
longitude = []

# get location raw data
for valor in regioesEncontradas:
    try:
        location = app.geocode(valor).raw
        latitude.append(location["lat"])
        longitude.append(location["lon"])
        i += 1
    
    except:
        nome_local = linha
        latitude.append('null')
        longitude.append('null')
        i += 1
```

```python
# Preparando as variáveis de latitudes e longitudes
latitude = np.array(latitude)
longitude = np.array(longitude)
print(latitude)
print(longitude)
```

```python
# Criando o Dicionário que auxiliará a colocação das coordenadas na respectiva localização Completa
dictLat = dict(zip(regioesEncontradas, latitude))
dictLon = dict(zip(regioesEncontradas, longitude))
print(dictLat)
print(dictLon)
```

```python
# Utilizando o dicionário para criar a coluna com a respectiva latitude
lat = []
for valor in cidades['RegiaoComp']: # Utilizando a nova coluna completa
    for c, v in dictLat.items():
        if c == valor: #regiosEncontradas[0].split(",")[0] # Esse trexo pertence a versão 2.0, aqui será verificado utilizando a coluna criada com o endereço completo
            lat.append(v)
            break #Suspensão em caso de nomes iguais - Se este trexo precisar ser utilizado não é bom
cidades['Latitude'] = lat
```

```python
# Utilizando o dicionário para criar a coluna com a respectiva latitude
lon = []
for valor in cidades['RegiaoComp']:
    for c, v in dictLon.items():
        if c == valor:
            lon.append(v)
            break
cidades['Longitude'] = lon
```
